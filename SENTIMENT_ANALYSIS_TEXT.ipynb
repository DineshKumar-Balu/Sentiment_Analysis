```python
    import pandas as pd
    import trafilatura
    import string
    import re
    import csv
    df=pd.read_excel(r"/content/drive/MyDrive/assignment/Output Data Structure.xlsx")
    get=df.URL
    number=37
    finial_list=[]
    for link in get:
        [Total_positive,Total_negative,Polarity_Score,Subjectivity_Score,Fog_Index,Percentage_of_Complex_words,Average_Sentence_Length,count_complex_words,Total_words_after_remove,syllable_per_word,c,Average_word_length]=['','','','','','','','','','','','']
        flag=100
        from urllib.request import urlopen
        from urllib.error import *
        try:
            html = urlopen(link)
        except HTTPError as e:
            flag=1000
        except URLError as e:
            flag=1000
        else:
            active_link=link
        if(flag==100):
            my_file=trafilatura.fetch_url(active_link)
            Exact_file=trafilatura.extract(my_file)
            total_no_sen=Exact_file.count('.')
            Main_file=Exact_file.replace('\n',' ')
            new_string = Exact_file.translate(str.maketrans('','',string.punctuation))

            Exact_file=new_string.replace('“','')
            Exact_file=Exact_file.replace('”','')

            pattern =r'[0-9]'
            newest_str=re.sub(pattern, '',Exact_file)
            newest_str=newest_str.replace('’s','')
            app=(newest_str.split())
            my_original=[]
            for i in app:
                   my_original.append(i.lower())
            with open('/content/drive/MyDrive/Stop_Words (1).txt') as f:
              new_stopwords=f.read()
            new=new_stopwords.split(' ')
            mynewstop_low=[]
            for i in new:
                mynewstop_low.append(i.lower())
            my_li=[]
            for u in my_original:
                if u not in mynewstop_low:
                    my_li.append(u)
            total_no_words=len(my_original)
            Total_words_after_remove=len(my_li)
            with open('/content/drive/MyDrive/Negative_Words (1).txt')as f:
              neg_wod=f.read()
              neg=neg_wod.split(' ')
            with open('/content/drive/MyDrive/Positive_Words (1).txt')as f:
              pos_wod=f.read()
              pos=pos_wod.split(' ')
            count_positive=[]
            count_negative=[]
            for i in my_li:
                if i in pos:
                    count_positive.append(i)
                if i in neg:
                     count_negative.append(i)
                Total_positive=(len(count_positive))
            Total_negative=len(count_negative)
            Polarity_Score = (Total_positive - Total_negative)/ ((Total_positive + Total_negative) + 0.000001)
            Subjectivity_Score = (Total_positive + Total_negative)/ ((Total_words_after_remove) + 0.000001)
            Average_Sentence_Length =total_no_words / total_no_sen
            total_syllable_count=[]
            for word in my_original:
                def syllable_count(word):
                    count = 0
                    vowels = "aeiouy"
                    if word[0] in vowels:
                        count += 1
                    for index in range(1, len(word)):
                        if word[index] in vowels and word[index - 1] not in vowels:
                             count += 1
                    if word.endswith("e"):
                        count -= 1
                    if (word.endswith("es") and word.endswith("ed")):
                        count -= 1
                    if count == 0:
                        count += 1
                    return count
                count_word=syllable_count(word)
                total_syllable_count.append(count_word)
            count_complex_words=0
            for i in total_syllable_count:
                if i>2:
                    count_complex_words=count_complex_words+1
            syllable_per_word=(sum(total_syllable_count)/total_no_words)
            Percentage_of_Complex_words =  (count_complex_words / total_no_words)
            Fog_Index = 0.4 * (Average_Sentence_Length + Percentage_of_Complex_words)
            Average_Number_of_Words_Per_Sentence =  total_no_words /total_no_sen
            c=0
            personal_pronouns=['we', 'i','us','ours','my']
            for k in my_original:
                if k in personal_pronouns:
                    c=c+1
            Total_char=[]
            for i in my_original:
                Total_char.append(len(i))
                Average_word_length=sum(Total_char)/total_no_words

        small_list=[number,link,Total_positive,Total_negative,Polarity_Score,Subjectivity_Score,Average_Sentence_Length,Percentage_of_Complex_words*100,Fog_Index,Average_Sentence_Length,count_complex_words,Total_words_after_remove,syllable_per_word,c,Average_word_length]
        finial_list.append(small_list)
        number=number+1
    Finial_output=pd.DataFrame(finial_list,columns=['URL ID','URL','POSITIVE SCORE','NEGATIVE SCORE','POLARITY SCORE','SUBJECTIVITY SCORE','AVG SENTENCE LENGTH','PERCENTAGE OF COMPLEX WORDS','FOG INDEX','AVG NUMBER OF WORDS PER SENTENCE','COMPLEX WORD COUNT','WORD COUND','SYLLABLE PER WORD','PERSONAL PRONOUNS','AVG WORD LENGTH'])





```
